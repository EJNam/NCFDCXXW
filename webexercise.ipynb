```python
import re
import time
fdcfile = open(r'F:\企业数据\房信网\南昌房地产信息.csv', 'r', encoding='UTF-8')  #以只读的方式打开本地文件
fdcinfo = fdcfile.read()
lst = fdcinfo.split('\n')
path1709 = 'F:\企业数据\path1709.txt'
count = 0
```

```python
for i in range(len(lst)):
#    print(i)
    d1709 = re.findall(r",[\d]*.[\d]*d18/02d1",str(lst[i]))
    with open(path1709, 'a', encoding='utf-8') as f:
        f.write( str(d1709) + '\n')
        count = count + 1
        print('\r当前速度: {:.2f}%'.format(count*100/len(lst)),end='')
    time.sleep(0.01)
```

```python
def findMinAndMax(L):
    if(len(L) == 0):
        return (None, None)
    else:

        max = L[0]
        min = L[0]

        for value in L:
            if value > max:
                max = value
            if value < min:
                min = value

    #返回
    return (min,max)
if findMinAndMax([]) != (None, None):
    print('测试失败!')
elif findMinAndMax([7]) != (7, 7):
    print('测试失败!')
elif findMinAndMax([7, 1]) != (1, 7):
    print('测试失败!')
elif findMinAndMax([7, 1, 3, 9, 5]) != (1, 9):
    print('测试失败!')
else:
    print('测试成功!')
```
